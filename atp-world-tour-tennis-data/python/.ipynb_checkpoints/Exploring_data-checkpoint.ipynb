# Import data

```python
# Function to import <2017 data and append 2017 data
def import_tennisdata(pf_data, pf_header, pf_data_2017=None):
    # Import column titles as list
    header_txt = pf_header
    with open(pf_header, "r") as f:
        pf_header = f.readlines()
        pf_header = [header.rstrip("\n") for header in pf_header]
    
    # Import <2017 data into dataframe
    df = pd.read_csv(pf_data, names = pf_header)
    
    # Import and append 2017 data
    if pf_data_2017 is not None:
        df_2017 = pd.read_csv(pf_data_2017, names = pf_header)
        df = df.append(df_2017)
    
    return df
```

```python
header = "csv\\3_match_stats\\match_stats_column_titles.txt"
data = "csv\\3_match_stats\\match_stats_1991-2016_UNINDEXED.csv"
data_2017 = "csv\\3_match_stats\\match_stats_2017_UNINDEXED.csv"
match_stats = import_tennisdata(data, header, data_2017)

match_stats.head()
```

# Merge data

```python
match_data = pd.merge(match_stats, match_scores, how="inner", on="match_id", suffixes=("", "_y"))
#Remove duplicate columns
col_to_delete = [col for col in match_data.columns if '_y' in col]
match_data.drop(labels=col_to_delete, axis=1, inplace=True)
match_data.columns
```

```python
#Drop duplicate rows
match_data.drop_duplicates(subset="match_stats_url_suffix", inplace=True)
match_data.shape
```

# Extract some stats

### Extract win ratio per player

```python
#Calculate win ratio per player
wins = match_data["winner_name"].value_counts()

losses = match_data["loser_name"].value_counts()
total_matches = wins + losses
#Keep only players with matches > 50 and drop NaN's
total_matches = total_matches[total_matches > 50].dropna()
win_ratio = wins / (total_matches)
win_ratio = win_ratio.dropna()
```

```python
win_ratio.sort_values(ascending=False)
```

### Extract yearly win ratio per player and plot

```python
match_data["year"] = match_data["match_id"].apply(lambda x: str(x)[0:4])
```

```python
win_count = match_data.groupby(by=["winner_name", "year"]).count()
win_count = win_count["match_id"].unstack()
loss_count = match_data.groupby(by=["loser_name", "year"]).count()
loss_count = loss_count["match_id"].unstack()
win_ratio_yearly = win_count/(loss_count + win_count)
win_ratio_yearly
```

```python
win_ratio_yearly.loc["Roger Federer"]
```

```python
#Plot the win/loss data for certain players -- barchart
import matplotlib.pyplot as plt
from numpy import arange

fig = plt.figure(figsize=(16,10))
ax = fig.add_subplot(1,1,1)
players = ["Roger Federer", "Rafael Nadal", "Novak Djokovic"]
x_offset = -0.2
for player in players:
    ax.bar(height=win_count.loc[player],
           x=arange(len(win_count.loc[player]))-x_offset,
           width=0.2,
           align="center",
           edgecolor="black",
           linewidth=0.1
           )
    x_offset += 0.2
    
xlabels = win_count.columns
ax.set_xticks(range(0, len(xlabels)))
ax.set_xticklabels(xlabels, rotation = 45, size=14)
ax.set_xlim(left=(1997-1991))
plt.show()
```

```python
#Plot the win/loss data for certain players -- lineplot
import matplotlib.pyplot as plt
from numpy import arange

fig = plt.figure(figsize=(16,14))
players = ["Roger Federer", "Rafael Nadal", "Novak Djokovic"]
colors = ["red", "#f4c441", "#41a3f4"]

ax1 = fig.add_subplot(3,1,1)
for i, player in enumerate(players):
    ax1.plot(win_count.loc[player],
            color=colors[i],
           linewidth=2,
           dashes=(3,2),
           marker="o")

ax2 = fig.add_subplot(3,1,2)
for i, player in enumerate(players):
    ax2.plot(loss_count.loc[player],
            color=colors[i],
           linewidth=2,
           dashes=(3,2),
           marker="o")    
    
ax3 = fig.add_subplot(3,1,3)
for i, player in enumerate(players):
    ax3.plot(win_ratio_yearly.loc[player],
            color=colors[i],
           linewidth=2,
           dashes=(3,2),
           marker="o")    
    
xlabels = win_count.columns

axes = [ax1, ax2]
for ax in axes:
    ax.set_xticks(range(0, len(xlabels)))
    ax.set_xlim(left=(1997-1991))
    ax.set_ylim(bottom=0, top=100)
    ax.tick_params(labelsize=14)
    ax.grid(b=True, color="#d9d9d9", linewidth=1.5)
    ax.set_xticklabels([])
    
ax3.set_xticklabels(xlabels, rotation = 45)
ax3.tick_params(labelsize=14)
ax3.grid(b=True, color="#d9d9d9", linewidth=1.5)
ax3.set_xlim(left=(1997-1991))
ax3.set_ylim(0.2, 1)

ax2.set_ylim(0,30)

ax1.set_title("Wins", size=20, fontweight="bold")
ax2.set_title("Losses", size=20, fontweight="bold")
ax3.set_title("Win ratio", size=20, fontweight="bold")
ax1.legend(loc="upper left", fontsize="x-large")
plt.show()
```

# Extract tiebreaks stats per player

```python
#Group tiebreak wins and losses for each winner of the match
winner_tb = match_data.groupby(by=["winner_name"]).sum()
winner_tb = winner_tb[["winner_tiebreaks_won", "loser_tiebreaks_won"]]

#Group tiebreak wins and losses for each winner of the match
loser_tb = match_data.groupby(by=["loser_name"]).sum()
loser_tb = loser_tb[["winner_tiebreaks_won", "loser_tiebreaks_won"]]

#Sum up data to get total wins/losses per player
tb_data2 = pd.DataFrame(index=winner_tb.index)
tb_data2["tiebreaks_won"] = winner_tb["winner_tiebreaks_won"] + loser_tb["loser_tiebreaks_won"]
tb_data2["tiebreaks_lost"] = winner_tb["loser_tiebreaks_won"] + loser_tb["winner_tiebreaks_won"]
#Create column with win ratio
tb_data2["win_ratio"] = tb_data2["tiebreaks_won"]/(tb_data2["tiebreaks_won"]+tb_data2["tiebreaks_lost"])

tb_data_elite = tb_data2[tb_data2["tiebreaks_won"] >= 50]
tb_data_elite.sort_values("win_ratio", ascending=False)


```
