### Import match data

```python
import requests
import json

response = requests.get("http://worldcup.sfg.io/matches", params="?by_date=desc")
response.status_code

match_data_json = response.json()
```

```python
#Results as list of dictionaries
match_data_json[0]
```

### Create dataframes

Create dataframe of match_data

```python
import pandas as pd

match_data = pd.DataFrame.from_dict(match_data_json)
match_data = match_data[match_data["status"] != "future"] #Keep out future matches
match_data
```

```python
#Function that creates dataframe from dictionary and merges with fifa_id
def join_df_with_fifa_id(dictionary, col_prefix):
    df = pd.DataFrame.from_dict(dictionary)
    df.columns = [col_prefix + "_" + str(col) for col in df.columns]
    df = df.join(match_data["fifa_id"], how="left")
    return df
    
```

Create dataframes of dictionaries within match_data DataFrame

```python
away_team_list = [match["away_team"] for match in match_data_json if match]
#Filter out the None's
away_team_list = list(filter(lambda x: x["code"] != "TBD", away_team_list))

away_team = join_df_with_fifa_id(away_team_list, "a")

print(away_team.shape)
away_team.head()
```

```python
away_team_events_list = []
for i, row in match_data.iterrows():
    for item in row["away_team_events"]:
        dic = {"id": item["id"],
              "type_of_event": item["type_of_event"],
              "player": item["player"],
              "time": item["time"],
               "fifa_id": row["fifa_id"]
              }
        away_team_events_list.append(dic)

away_team_events = pd.DataFrame.from_dict(away_team_events_list)

print(away_team_events.shape)
away_team_events.head()

        
        
```

```python
away_stats_list = [match["away_team_statistics"] for i, match in match_data.iterrows()]
#Filter out the None's
#away_stats_list = list(filter(lambda x: x != None, away_stats_list))

away_stats = join_df_with_fifa_id(away_stats_list, "a")

print(away_stats.shape)
away_stats.head(5)
```

```python
home_team_list = [match["home_team"] for match in match_data_json]
#Filter out the None's
home_team_list = list(filter(lambda x: x["code"] != "TBD", home_team_list))

home_team = join_df_with_fifa_id(home_team_list, "h")

print(home_team.shape)
home_team.head()
```

```python
home_team_events_list = []
for i, row in match_data.iterrows():
    for item in row["home_team_events"]:
        dic = {"id": item["id"],
              "type_of_event": item["type_of_event"],
              "player": item["player"],
              "time": item["time"],
               "fifa_id": row["fifa_id"]
              }
        home_team_events_list.append(dic)

home_team_events = pd.DataFrame.from_dict(home_team_events_list)

print(home_team_events.shape)
home_team_events.head()
```

```python
home_stats_list = [match["home_team_statistics"] for i, match in match_data.iterrows()]
#Filter out the None's
#home_stats_list = list(filter(lambda x: x != None, home_stats_list))

home_stats = join_df_with_fifa_id(home_stats_list, "h")

print(home_stats.shape)
home_stats.head(5)
```

Merge all dataframes

```python
dfs = [away_team, away_stats, home_team, home_stats]
cols_to_drop = ["away_team", "away_team_statistics",
               "home_team", "home_team_statistics"]
match_data_full = match_data.drop(cols_to_drop, axis=1)
for df in dfs:
    match_data_full = match_data_full.merge(df, how="inner", on="fifa_id", suffixes=["","_x"])

#Remove overlapping columns
new_cols = [col for col in match_data_full.columns if "_x" not in col]
match_data_full = match_data_full[new_cols]

match_data_full.head()
```

```python
import re
import numpy as np

#Columns to keep and what function will be applied to that columns when grouping the data by country
cols_to_agg = {"goals": np.sum, 
               "attempts_on_goal": np.sum,
               "ball_possession": np.mean,
               "balls_recovered": np.sum,
               "blocked": np.sum,
               "clearances": np.sum,
               "corners": np.sum,
               "distance_covered": np.sum,
               "fouls_committed": np.sum,
               "num_passes": np.sum,
               "off_target": np.sum,
               "offsides": np.sum,
               "on_target": np.sum,
               "pass_accuracy": np.mean,
               "passes_completed": np.sum,
               "red_cards": np.sum,
               "tackles": np.sum,
               "tackles": np.sum,
               "woodwork": np.sum,
               "yellow_cards": np.sum
              }

#Sum the stats by country
a_cols_to_agg = {}
for key in cols_to_agg:
    a_cols_to_agg["a_" + key] = cols_to_agg[key]
a_country_data = match_data_full.groupby(["a_country"]).agg(a_cols_to_agg)
#Add columns with number of matches
a_country_data["a_matches"] = match_data_full.groupby(["a_country"]).size()
#Extract only columns with team stats
newcols = [col for col in a_country_data.columns if "a_" in col]
a_country_data = a_country_data[newcols]
#Remove the suffixes of columnnames to prepare for appending
a_country_data.rename(columns=lambda x: x[2:], inplace=True)

h_cols_to_agg = {}
for key in cols_to_agg:
    h_cols_to_agg["h_" + key] = cols_to_agg[key]
h_country_data = match_data_full.groupby(["h_country"]).agg(h_cols_to_agg)
h_country_data["h_matches"] = match_data_full.groupby(["h_country"]).size()
newcols = [col for col in h_country_data.columns if "h_" in col]
h_country_data = h_country_data[newcols]
h_country_data.rename(columns=lambda x: x[2:], inplace=True)

#Append home and away country stats in one dataframe
country_stats = h_country_data.append(a_country_data)
country_stats.reset_index(inplace=True)
cols_to_agg["matches"] = np.sum #Add number of matches to list of columns to aggregate
country_stats = country_stats.groupby(by="index").agg(cols_to_agg)
country_stats.head()


```

```python
import matplotlib.pyplot as plt

#Create fonts for plots
title_font = {"fontname": "Calibri", 
         "fontsize": 16,
         "weight": "heavy"
        }


annotate_font = {"fontname": "Verdana", 
         "fontsize": 10,
         "weight": "bold",
         "verticalalignment": "center"
        }

#Function that plots each column through a loop
def createPlots(cols):
    fig = plt.figure(figsize=(16,8))
    for i, col in enumerate(cols):
        ax = fig.add_subplot(3,2,i+1) #Change number of subplots if necessary
        x = country_stats[col].sort_values().tail()
        y = [country[0:3] for country in x.index]
        ax.barh(y, 
                 width=x,
                 height=0.6,
                 color="#ff4d4d",
                 edgecolor="black",
                 linewidth=1.3,
                )
        ax.set_title(re.sub("_"," ",col), **title_font)
        ax.tick_params(axis="y", labelsize=12)
        ax.tick_params(axis="x", labelsize=12)
        for value, country in zip(x, y):
            ax.annotate(str(value), xy=(value + x.max()/100, country), **annotate_font)
    plt.tight_layout()
    plt.show()
```

```python
#List the columns to plot
cols = [col for i, col in enumerate(country_stats.columns) if i<6]
createPlots(cols)
```
