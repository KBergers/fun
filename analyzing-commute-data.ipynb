First download Location History data dump in json from https://takeout.google.com/settings/takeout. The format of the dump is very simple, it’s a dict with locations key.

```python
# Parse the location data
import json
from datetime import datetime
from collections import namedtuple

Point = namedtuple('Point', 'latitude, longitude, datetime')


def read_points():
    with open('C:\\Users\\SWP679\\Documents\\Location History.json') as f:
        data = json.load(f)
        
    points = []
    for point in data['locations']:
        points.append(Point(
            point['latitudeE7'] / 10 ** 7,
            point['longitudeE7'] / 10 ** 7,
            datetime.fromtimestamp(int(point['timestampMs']) / 1000)
        ))
        
    return points
        
points = read_points()
```

```python
# Ignore entries before certain date and take out weekends
from itertools import takewhile

from_date = datetime(2018, 3, 1)
after_move = takewhile(lambda point: point.datetime >= from_date, points)

work_days = (point for point in after_move if point.datetime.weekday() < 5)
```

```python
# Defines the times in which you want to take the data
from_hour = 7
to_hour = 11

commute_time = (point for point in work_days if from_hour <= point.datetime.hour < to_hour)
```

```python
# Group everything by day
from itertools import groupby
by_days = groupby(commute_time, key=lambda point: point.datetime.date())
```

Select the last point at home and the first point at work for every day. A point is considered home or work if it’s distance from home or work is smaller than 50 meters. 

```python
from geopy.distance import geodesic
home = (51.205613, 4.397171)  # not really =)
work = (51.199814, 4.418470)

max_distance = 0.100

def last_at_home(points):
    result = None
    for point in points:
        if geodesic(home, point[:2]).km <= max_distance:
            result = point
    return result

def first_at_work(points, after):
    for point in points:
        if point.datetime > after.datetime and geodesic(work, point[:2]).km <= max_distance:
            return point
        
Commute = namedtuple('Commute', 'day, start, end, took')

def get_commute():
    commute = []
    for day, points in by_days:
        points = [*points][::-1] #Negative slicing: puts the bottom value at top

        start = last_at_home(points)
        if start is None:
            continue

        end = first_at_work(points, start)
        if end is None:
            continue

        commute.append(Commute(
            day, start.datetime, end.datetime, end.datetime - start.datetime,
        ))
    return commute

commutes = get_commute()
```

```python
# Plot graph of daily commute
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(16,10))
ax.plot([commute.day for commute in commutes],
        [commute.took.total_seconds() / 60 for commute in commutes])

ax.set(xlabel='day', ylabel='commute (minutes)',
       title='Daily commute')

ax.grid()
plt.show()
```
